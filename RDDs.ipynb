#### pyspark RDD practice 

```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('Read File')
sc = SparkContext.getOrCreate(conf=conf)
```

Read and split text file into RDD

```python
rdd = sc.textFile('./data_samples/quiz.txt')
```

```python
mappedRdd = rdd.flatMap(lambda x: x.split(" "))  #takes elements from inside sub-brackets
mappedRdd.collect()
```

filter with normal function 

```python
def filterAandC(x):
    if x.startswith('c') or x.startswith('a'):
        return False
    return True

filtrdRdd = mappedRdd.filter(filterAandC)
filtrdRdd.collect()
```

filter with lambda function

```python
filteredRddlambda = mappedRdd.filter(lambda x: not (x.startswith('c') or x.startswith('a')))
filteredRddlambda.collect()
```

group by

```python
mappedRdd.map(lambda x: (x, len(x))).groupByKey().mapValues(list).collect()
```

reduce by for aggregation

```python
reduced= mappedRdd.map(lambda x: (x, 1)).reduceByKey(lambda x,y: x+y)
```

```python
reduced.saveAsTextFile('/FileStore/tables/output/samle_words_output.txt')
```

```python
reduced = reduced.repartition(5)
```
